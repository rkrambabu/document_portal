RAG has 3 component:
* Data ingestion
* Data Retrieval
* Data Generation

Chunking will be performed on the text data type not on image or video
embedding means convert the chunk data into vector. Vector is nothing but embedding
audio and image can be converted into embedding
we can also convert the audio and video into text using some visper model then convert into embedding or vector
chunking size willbe experimental value or hyperparameter
chunking done on the large text data

If anyone ask the question then it will convert to vector then do the semantic serch and retrieve the data which is called retrieval

Data retrieval
Fetch the data from the vector DB
The query or question will be converted into embeding then perform the vector operation like similarity search then retrieve the relevant content or ranked result.
Fetch most 10 relevant result then again with fetched data we fetch most 3 relevant result which is called re-ranking.

after retrieval we will have the context, add the query and prompt then pass the LLM then LLM is going to generate our answer.
Here instead of asking direct query to LLM we are retrieving information from VDB which is called context then adding query along with prompt passing to LLM.

Data Generation
We collect the retrieved information based on the query from Vector DB which is also called filtered information. Here we have prompt with some additional information and query to pass along with then LLM generate the final answer.
